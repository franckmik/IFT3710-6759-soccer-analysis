```python
import torch
from torch.utils.data import DataLoader
from torchvision.transforms import transforms
from my_module import SoccerEventDataset, SoccerEventClassifier,evaluate_model

#Transformation to apply to the images
transform= transforms.Compose([
      transforms.Resize(256),
      transforms.CenterCrop(224),
      transforms.ToTensor(),
      transforms.Normalize(mean=[.485,.456,.406],std=[.229,.224,.225])
  ])

# runtime to evaluate model
device= torch.device('cuda' if torch.cuda.is_available() else 'cpu')

root='/content/drive/MyDrive/FootballAnalysis/Test' #root dir

batch_size=16

test_dataset= SoccerEventDataset(root_dir=root,transform=transform)

test_loader= DataLoader(test_dataset,batch_size=batch_size,shuffle=False)

model= SoccerEventClassifier().to(device)

model.load_state_dict(
    torch.load('/content/drive/MyDrive/FootballAnalysis/model_event_classifier.pth'
    ))
accuracy= evaluate_model(model=model,test_loader=test_loader, device=device)
```

```python
print(f'Accuracy: {accuracy}')
```
